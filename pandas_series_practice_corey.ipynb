# Filtering Data


```python
people={"first":["corey","jane","john"],
       "last":["Schafer","Doe","Doe"],
       "email":["corey@gmail.com","jane@gmail.com","john@gmail.com"]}
```

```python
df=pd.DataFrame(people)
```

```python
df
```

```python
df['last']=="Doe"  #gives series of true or false values meeting filter criteria
```

```python
filt=(df['last']=="Doe")
```

```python
filt
```

```python
df[filt]
```

```python
df[df['last']=="Doe"] #same as above
```

```python
df.loc[filt]
```

```python
df.loc[filt,"email"]
```

```python
filt=(df['last']=='Doe') & (df['first'] == 'john')
```

```python
df.loc[filt,"email"]
```

```python
df
```

```python
filt=(df['last']=='Schafer') |  (df['first'] == 'john')
```

```python
df.loc[filt,"email"]
```

```python
df.loc[~filt,"email"]
```
